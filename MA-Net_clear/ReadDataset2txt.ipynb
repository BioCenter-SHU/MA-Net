```python
import os
file_path = "/YOUR_PATH/MyDataset/0_single_plaque/video_302_frame_120/Mixed/"
# 当前文件夹
video_name_file_list = os.listdir(file_path)
# 所有视频
all_video_list = []
# 对当前文件夹遍历
for file in video_name_file_list:
    video_path = os.path.join(file_path, file)
    frame_name_list = os.listdir(video_path)
    ## 获取的list中是str，直接sort是按照str进行按位对比的，需要转化为int，再进行sort
    frame_name = []
    # （1）去掉.jpg
    for f in frame_name_list:
        frame_name.append(f[:-4])
    # print(frame_name)
    # (2)str转为int
    frame_name = list(map(int, frame_name))
    frame_name.sort()
    # (3)int转为str
    frame_name = list(map(str, frame_name))
    # 一个视频所有帧的list
    frame_list = []
    for frame in frame_name:
        frame_list.append('Mixed/'+ file + '/' + frame + '.jpg')
    # print(frame_list)
    all_video_list.append(frame_list)
# print(all_video_list)
f = open('frame_120/data_Mixed.txt', 'w')
for i in range(len(all_video_list)):
    f.write(str(all_video_list[i])+'|'+'2'+'\n')
f.close()
```

```python
# 计算标准差和均值
import torch
from data import dataset
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
import os
#要自己用的话记得改改路径哦！
path = "/YOUR_PATH/MyDataset/0_single_plaque/video_302_frame_120/" #我这里使用了绝对路径是因为我老是出错，搞得文件到处乱放
train_data = dataset.dataset_video(is_train=True, frame_path=path, root='frame_120/KFold_302_video.txt', clip_len=120)
train_loader = DataLoader(dataset=train_data, batch_size=1, shuffle=True, num_workers=1)
def get_mean_std_value(loader):
    batch = 1
    frame = 120
    channels_sum,channel_squared_sum,num_batches = 0,0,0

    for data,target in loader:
        nt, c, h, w = data.size()
        channels_sum += torch.mean(data,dim=[0,2,3])#shape [n_samples(batch),channels,height,width]
        #并不需要求channel的均值
        channel_squared_sum += torch.mean(data**2,dim=[0,2,3])#shape [n_samples(batch),channels,height,width]
        num_batches +=1

    # This lo calculate the summarized value of mean we need to divided it by num_batches

    mean = channels_sum/num_batches
    #这里将标准差 的公式变形了一下，让代码更方便写一点

    std = (channel_squared_sum/num_batches - mean**2)**5
    return mean,std

mean,std = get_mean_std_value(train_loader)
print('mean = {},std = {}'.format(mean,std))
```
